Creating a Workspace in Microsoft Fabric is quite straightforward — but there are a few important settings to get right so you can later assign capacities, manage access, and publish content.

Here’s a step-by-step guide:

# 1. Sign in to Microsoft Fabric
Go to https://app.fabric.microsoft.com (same as the Power BI Service portal).

Sign in using your Microsoft 365 / organizational account that has the Fabric license or trial enabled.



# 2. Open the Workspace Creation Menu
On the left navigation pane, click Workspaces.

At the top right, click + New workspace.

# 3. Fill in Workspace Details
You’ll see the Create a workspace panel. Fill in the following:
| Field                        | Description                                                        |
| ---------------------------- | ------------------------------------------------------------------ |
| **Name**                     | The display name of your workspace (e.g., *Sales Analytics 2025*). |
| **Description** *(optional)* | Purpose or notes for the workspace.                                |
| **Image** *(optional)*       | Workspace icon for easy identification.                            |


# 4. Assign a Capacity
Fabric workspaces require being assigned to a capacity to use all Fabric workloads.

In Advanced settings:

Select Assign to capacity.

Choose your available Fabric capacity (F SKU, P SKU, or trial capacity).

If you don’t assign capacity, the workspace will behave like a normal Power BI workspace without Fabric features.

# 5. Configure Permissions
Under Workspace access:

Add Admins, Members, Contributors, Viewers by entering their names or email addresses.

Choose the right role:

- Admin – Full control, manage settings, delete workspace.

- Member – Edit, publish, and share content.

- Contributor – Create and update items, but no sharing.

- Viewer – Read-only access.

Best practice:

Use security groups instead of individuals for easier management.

Keep Admin count low to avoid accidental workspace deletion.


# 6. Save the Workspace
Click Save or Create.

The workspace appears under Workspaces in your navigation pane.

✅ You can now:

Create and publish Data Warehouses, Lakehouses, Notebooks, Pipelines, Reports, etc., inside this workspace.

Manage permissions and settings later via Workspace settings.

# 7. Advanced Settings
Click Advanced to expand options:

Assign to capacity
Toggle ON to enable Fabric features.

Choose a Capacity from the dropdown.

If your account has only one capacity, it will auto-select.

If you skip this, the workspace will be limited to standard Power BI features (no Fabric Lakehouse/Warehouse).

OneLake integration
Fabric automatically integrates your workspace with OneLake storage.

All data items (Lakehouse, Warehouse, etc.) will store files in the OneLake folder tied to this workspace.

# 8. Post-Creation Management
After the workspace is created, you can:

### Change Settings

Go to the workspace → Settings (gear icon in top right).

Update:

- Name, description, image
- Capacity assignment
- Access permissions
- Default storage format

### Add Fabric Items
Inside the workspace, click + New to add:

- Lakehouse
- Data Warehouse
- Notebooks
- Pipelines
- Reports & Dashboards
- KQL Databases
- Eventstreams

### Monitor Usage
Go to Settings → Capacity metrics to see:

- Storage used
- Query performance
- Refresh schedules
